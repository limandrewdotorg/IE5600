### Difficulty Gap
Firstly observe that the new problem can be added in one of the largest gap, otherwise the new list's maximum gap will not be reduced.

Then observe that the problem can be added in the middle to minimize the gap as much as possible. Note that the gap may not be even.

Then calculate the maximum gap of the new list. Remark that don't forget the second largest gap in the original list. 

```python
import math
 
t= int(input())
for i in range(t):
    n = int(input())
    tlist = [int(j) for j in input().split()]
    gap = [tlist[k+1]-tlist[k] for k in range(n-1)]
    gap.sort()
    max_gap=  math.ceil(gap[-1]/2)
    if n<=2:
        print(max_gap)
    else:
        print(max(max_gap,gap[-2]))

```

### Problem Selection
This problem can be solved by greedy method. Since the number of problem should be as large as possible, the problem with smaller difficulty should be added first. 


```python
test = int(input())
for _ in range(test):
    n,k = map(int,input().split())
    a = list(map(int,input().split()))
    a.sort()
    a.reverse()
    cnt = 0
    while len(a) and k>=a[-1]:
        cnt+=1
        k-=a[-1]
        a.pop()
    print(cnt)
    
```

### Number Modification
It can be proved that it is always possible.

Firstly all negative number will be replaced to a non-negative number in $O(n)$ operation if it has negative number.

Then all $0$ will be replaced to a positive number in $O(n)$ if it has 0. 

Therefore we can observe that after $O(n)$ operation, all number will be strictly positive. 
Moreover, in any of the following operation all number will still be strictly positive as long as the maximum is minimum is different. 

We can also observe that one operation either reduce the maximum itself, or the number of maximum. This means if the maximum is $V$ after all non-positive number is replaced, we can run at most $nV$ operation to make all numer equal, otherwise contradict with the conclusion that all number is sitrctly positive after each iteration. 

This $V$ is not to large due to our setting. Therefore, a simple simulation algorithm should pass this problem. 


```python
test = int(input())
for _ in range(test):
    
    n = int(input())
    a = list(map(int,input().split()))
    while(max(a)>min(a)):
        ma = max(a)
        mi = min(a)
        a.remove(ma)
        a.remove(mi)
        a.append(ma-mi)
        a.append(ma-mi)
    print(a[0])
```

### The Adventure - Pathway
Binary search the strength.

To check whether a strenth $x$ is sufficient, a DFS/BFS should be enough. 

One may also solve it with a minimum spanning tree algorithm or a modified single source shortest path algorithm.

```python
import queue
import time


def solve():
    n,m = list(map(int, input().split()))
    s = list(map(int, input().split()))
    edge = {i:{} for i in range(1,n+1)}
    for _ in range(m):
        a,b = list(map(int, input().split()))
        if a==b:
            continue
        v = max(s[a-1],s[b-1])
        if a in edge[b]:
            if v <= edge[a][b]:
                continue

        edge[a][b]=v
        edge[b][a]=v
    
    if n==2: return 0

    def find(k):
        q = queue.Queue()
        q.put(1)
        vis = [0 for i in range(n)]
        while not q.empty():
            a = q.get()
            vis[a-1]=1
            neighbor = [b for b in edge[a] if (edge[a][b]<=k and vis[b-1]==0)]
            for b in neighbor:
                q.put(b)            
        
        return vis[n-1]==1

    k_min = 0
    k_max = max(s)
    mid = (k_min+k_max)//2

    while k_min<k_max-1:
        mid = (k_min+k_max)//2
        if(find(mid)):
            k_max = mid
        else:
            k_min = mid+1

    if(find(k_min)):
        return k_min
    return k_max
    


time_start = time.time()
t = int(input())

for _ in range(t):
    print(solve())


time_end = time.time()
# print('time cost',time_end-time_start,'s')
```

### Photo
Backtracking problem

Take care of the equivalence condition: Two arrangements are considered as different when the two heights of any position are different.

```python
test = int(input())
for _ in range(test):
    n = int(input())
    l = list(map(int,input().split()))
    
    dic = {}
    for u in l:
        if u not in dic: dic[u] = 0
        dic[u]+=1
    
    cnt = 0
    g = [[0 for j in range(n)] for i in range(n)]
    
    def f(x):
        global cnt
        if x==n*n:
            cnt+=1
            return
        r = x//n 
        c = x%n
        for u in dic:
            if (r==0 or u >= g[r-1][c]) and (c==0 or u >= g[r][c-1]) and dic[u]>0:
                
                g[r][c] = u
                dic[u]-=1
                f(x+1)
                dic[u]+=1
                
    f(0)
    
    print(cnt)
            
```

### The Adventure - Pizza
Firstly obtain the size of each slides before add additional $m$ cuts. This can be done by sort $x$.

Denote the size of each slides as $b_i$. 

Denote the final size of the slides after $m$ cuts to $v$. Then $m = \frac{360}{v} - n$. Therefore, $v$ should be as large as possible, while $\frac{b_i}{v}$ is an integer hold for all $i$. 

One may guess $v$ to be an integer and try all possible value from $1$ to $360$.

To prove $v$ to be an integer, one may first observe that $v$ should be rational, then the prove follow the Euclidean Algorithm (GCD), and therefore this problem can also be solved in this way. 

```python
import math

t = int(input())

for _ in range(t):
    n = int(input())
    x = list(map(int, input().split()))
    x = sorted(x)
    angle = []
    
    for i in range(1, n):
        angle.append(x[i] - x[i - 1])
    angle.append(360 - x[-1] + x[0])
    
    ans = math.gcd(angle[0], angle[1])
    for i in range(2, len(angle)):
        ans = math.gcd(ans, angle[i])
    
    res = 360 // math.gcd(360, ans) - n
    
    print(res)
```
